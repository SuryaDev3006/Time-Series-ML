```python
# %%
import os
import sys
from pathlib import Path

# ── Project root ───────────────────────────────────────────────────────────────
# This file lives in notebooks/. Adjust ROOT if you run from elsewhere.
__file__ = os.path.abspath("00_setup.ipynb")
ROOT = Path(__file__).parent.parent
sys.path.insert(0, str(ROOT))

# ── Create all directories ─────────────────────────────────────────────────────
DIRS = [
    "data/raw", "data/processed", "data/features", "data/splits",
    "artifacts/plots", "artifacts/metrics", "artifacts/models",
    "configs/data", "configs/features", "configs/models",
    "configs/evaluation", "configs/monitoring",
    "notebooks", "scripts",
]
for d in DIRS:
    (ROOT / d).mkdir(parents=True, exist_ok=True)
```

```python
import mlflow

USE_DAGSHUB = True

if USE_DAGSHUB:
    import dagshub

    dagshub.init(
        repo_owner="suryateja6969",
        repo_name="Time-Series-ML",  # change if your DagsHub repo uses a different name
        mlflow=True,
    )
else:
    TRACKING_URI = str(ROOT / "mlruns")
    mlflow.set_tracking_uri(TRACKING_URI)

EXPERIMENT_NAME = "walmart-m5"
mlflow.set_experiment(EXPERIMENT_NAME)
```
